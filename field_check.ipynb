```python
import pandas as pd
import numpy as np
```

```python
origin_data = pd.read_csv("data/lending_club_2020_train.csv")
origin_data.head()
```

```python
lst = []
with open("find_list.txt", mode='r') as f:
    while 1:
        line = f.readline().strip('\n')
        if not line: break
        lst.append(line)
print(len(lst), lst)
```

```python
fields = origin_data.columns
```

```python
def check_fields(lst):
    for index, field in enumerate(lst):
        try:
            origin_data[field]
            print(f"=========================================={index, field}==========")
            print("======value_count======")
            print(origin_data[field].value_counts())
            print("======Null percentage======")
            print((origin_data[field].isnull().sum()*100)/len(origin_data))
            print("======NaN percentage======")
            print((origin_data[field].isna().sum()*100)/len(origin_data))
            print("======data head======")
            print(origin_data[field].head())
        except KeyError as ke:
            print(ke, field)
```

```python
check_fields(lst)
```
